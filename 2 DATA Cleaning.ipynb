# Lemmatization

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
word_net = WordNetLemmatizer()        #creating word net lemmatizing object


def pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:         
        return None
```

```python
def lemmatizing(i):
    words_split = i.split()   #splitting words
    
    no_stopwords = [i for i in words_split if i not in stopwords.words("English")]  #removing stopwords
     #Giving part of speech to each word
    no_stopwords = nltk.pos_tag(no_stopwords)      
    
    List=[]                   
    
    for i in no_stopwords:
        a=(i[0],pos_tagger(i[1]))
        List.append(a)


    # apply pos tag and lemmatization
    list_of_strings = []
    for i in List:
        try: 
            list_of_strings.append(word_net.lemmatize(i[0],i[1]))
        except:
            list_of_strings.append(word_net.lemmatize(i[0]))
    return " ".join(list_of_strings)


x = "Hinge is good and phone is handy CG glass Victus is good Display is good No fast charging"
lemmatizing(x)
```

```python
import time
START = time.time()
df['cleaned_review'] = df['cleaned_review'].apply(lemmatizing)
STOP = time.time()
print((STOP - START)/60)
```
