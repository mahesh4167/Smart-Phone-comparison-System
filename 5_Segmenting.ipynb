```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv(r"/content/drive/MyDrive/DATASETS/DATA.csv")
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head(5)
```

```python
def extract_company_name(text):
   return (re.match(r'^\w+',text).group(0)).upper()
```

```python
df['Brand_name'] = df['model_name'].apply(extract_company_name)
```

**For transformers we dont need the cleaned review as it is lemmatized , so we drop that column**

```python
df.drop('cleaned_review',axis=1,inplace=True)
```

```python
# Rearranging columns
new_order = ['Brand_name','model_name','price','review','rating_by_reviewer']
df = df[new_order]
df.head(5)
```
